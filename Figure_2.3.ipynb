```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- Figure 2.3 --- #

# Defining the visible energy range [MeV]
ev = np.linspace(1,10,101)

# Computing the detector energy resolution
sigma = ev*np.sqrt(0.0261**2/ev + 0.0064**2 + 0.012**2/ev**2)

#Plotting
plt.figure(dpi = 300)
plt.plot(ev,100*sigma, '--', c = 'black')
plt.xticks([1,2,4,6,8,10])
plt.ylabel(r'$\sigma_{E_V} \ [\%]$')
plt.xlabel(r'$E_V \ [\text{MeV}]$')
plt.grid(alpha = 0.5, ls = ':')
plt.tight_layout()
plt.show()
```
