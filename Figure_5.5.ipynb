```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.gridspec as gridspec
```

```python
# --- Figure 5.5 --- #

# Uses the figures generated in the file 'Individual_sens_plots.ipynb' #

image_paths = ['NO_sens.png', 'IO_sens.png', 'Both_sens.png']
images = [mpimg.imread(path) for path in image_paths]

fig = plt.figure(figsize=(16, 10), dpi=300, constrained_layout=True)
gs = gridspec.GridSpec(2, 2, figure=fig, height_ratios=[1, 1], width_ratios=[1, 1], wspace = 0)

ax1 = fig.add_subplot(gs[0, 0])
ax1.imshow(images[0])
ax1.axis('off')

ax2 = fig.add_subplot(gs[0, 1])
ax2.imshow(images[1])
ax2.axis('off')

ax3 = fig.add_subplot(gs[1, :])
ax3.imshow(images[2])
ax3.axis('off')

plt.show()
```
