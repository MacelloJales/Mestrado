```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- Figure 3.2 --- #

# Physical constants
Enu = 5 # Fixed electron antineutrino energy [MeV]
Me = 0.511 # Electron mass [MeV]
Mn = 939.565 # Neutron mass [MeV]
Mp = 938.272 # Proton mass [MeV]
M = (Mn+Mp)/2 # Average nucleon mass [MeV]
Delta = Mn - Mp # Mass diference [MeV]

# Zeroth order positron energy and velocity (arXiv:hep-ph/9903554)
Ee0 = Enu - Delta 
ve0 = np.sqrt(1 - Me**2/Ee0**2)

# Defining the scattering angle range
theta = np.deg2rad(np.arange(0,361))

# Computing the first-order positron energy [MeV] (arXiv:hep-ph/9903554)
y = (Delta**2 - Me**2)/2
E = Ee0*(1 - (Enu/M)*(1 - ve0*np.cos(theta))) - y/M 

# Plotting
plt.figure(dpi = 300)
plt.xlabel(r'$\cos\theta$')
plt.ylabel(r'$E_{e^+}^{(1)}(5 \ \text{MeV}, \cos\theta) \ [\text{MeV}]$')
plt.plot(np.cos(theta), E, ':', c = 'black')
plt.grid(alpha = 0.5, ls = ':')
plt.tight_layout()
plt.show()
```
