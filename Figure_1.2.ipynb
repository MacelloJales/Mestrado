```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- Figure 1.2 --- #

# Computes the 3-flavor electron antineutrino survival probability in vacuum, considering the Normal Ordering.
def Prob_NO(Enu, L):

    # Oscillation Parameters NO best-fit (arXiv:2410.05380)
    dm21 = 7.49e-5
    dm31 = 2.534e-3
    t12 = 33.68
    t13 = 8.52
    t23 = 48.5
    
    # Defining Δm²₃₂
    dm32 = dm31 - dm21
    
    # Converting mixing angles to radians
    t12 = np.deg2rad(t12)
    t13 = np.deg2rad(t13)
    t23 = np.deg2rad(t23)
    
    # Sines and cosines of the mixing angles 
    s2_t12 = np.sin(t12)**2
    c2_t12 = np.cos(t12)**2
    s2_2t12 = np.sin(2.0*t12)**2
    s2_2t13 = np.sin(2.0*t13)**2
    c4_t13 = np.cos(t13)**4
    
    # Oscillation phases in vacuum
    d21 = 1.27e3*dm21*L/Enu
    d31 = 1.27e3*dm31*L/Enu
    d32 = 1.27e3*dm32*L/Enu
    s2_d21 = np.sin(d21)**2
    s2_d31 = np.sin(d31)**2
    s2_d32 = np.sin(d32)**2
    
    # Calculating
    prob_v = 1.0 - s2_2t12*c4_t13*s2_d21 - s2_2t13*(c2_t12*s2_d31 + s2_t12*s2_d32)
    
    return prob_v

# Computes the 3-flavor electron antineutrino survival probability in vacuum, considering the Inverted Ordering.
def Prob_IO(Enu, L):
    
    # Oscillation Parameters IO best-fit (arXiv:2410.05380)
    dm21 = 7.49e-5
    dm31 = -2.51e-3
    t12 = 33.68
    t13 = 8.58
    t23 = 48.6
    
    # Defining Δm²₃₂
    dm32 = dm31 - dm21
    
    # Converting mixing angles to radians
    t12 = np.deg2rad(t12)
    t13 = np.deg2rad(t13)
    t23 = np.deg2rad(t23)
    
    # Sines and cosines of the mixing angles
    s2_t12 = np.sin(t12)**2
    c2_t12 = np.cos(t12)**2
    s2_2t12 = np.sin(2.0*t12)**2
    s2_2t13 = np.sin(2.0*t13)**2
    c4_t13 = np.cos(t13)**4
    
    # Oscillation phases in vacuum
    d21 = 1.27e3*dm21*L/Enu
    d31 = 1.27e3*dm31*L/Enu
    d32 = 1.27e3*dm32*L/Enu
    s2_d21 = np.sin(d21)**2
    s2_d31 = np.sin(d31)**2
    s2_d32 = np.sin(d32)**2
    
    # Calculating
    prob_v = 1.0 - s2_2t12*c4_t13*s2_d21 - s2_2t13*(c2_t12*s2_d31 + s2_t12*s2_d32)
    
    return prob_v
```

```python
# Defining energy range to plot
E = np.linspace(1.6,10,501) # Reactor neutrino energies [MeV]
L = 52.5 # Fixed baseline [km]

# Computing the survival probabilities for each ordering
v_NO = np.array(Prob_NO(E,L))
v_IO = np.array(Prob_IO(E,L))

# Plotting
plt.figure(dpi = 300, figsize=(10,5))
plt.plot(E, v_NO, 'b', label = 'Normal Ordering')
plt.plot(E, v_IO, 'r', label = 'Inverted Ordering')
plt.xticks(fontsize = 11.5)
plt.yticks(fontsize = 11.5)
plt.xlabel('Electron Antineutrino Energy [MeV]', fontsize = 13)
plt.ylabel(r'$P_{\bar{\nu}_e \rightarrow \bar{\nu}_e}$', fontsize = 15)
plt.minorticks_on()
plt.legend(fontsize = 13)
plt.tight_layout()
plt.show()
```
