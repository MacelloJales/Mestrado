```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- Figure 3.4 --- #

# Defining the visible energy selected values [MeV]
Ev = [3,4,5,6,7]

# Defining the reconstructed energy range [MeV]
E = np.linspace(2.5,7.5,10001)

# List to store the response functions for each visible energy value
curves = []

# Loop over the selected visible energies
for i in Ev:
    
    # Computing the energy resolution
    sigma = i*np.sqrt(0.0261**2/i + 0.0064**2 + 0.012**2/i**2)
    
    # Computing the detector response function
    resol = np.exp(-0.5 * ((E - i) / sigma) ** 2.)/np.sqrt(2.*np.pi)/sigma
    
    # Storing the computed response curve
    curves.append(resol)

# Plotting
plt.figure(dpi = 300)
plt.plot(E, curves[0], lw = 1, label = r'$E_V = 3 \ \text{MeV}$')
plt.plot(E, curves[1], label = r'$E_V = 4 \ \text{MeV}$')
plt.plot(E, curves[2], label = r'$E_V = 5 \ \text{MeV}$')
plt.plot(E, curves[3], label = r'$E_V = 6 \ \text{MeV}$')
plt.plot(E, curves[4], label = r'$E_V = 7 \ \text{MeV}$')
plt.ylim(0.02, 8.6)
plt.xlabel(r'$\bar{E}_V \ [\text{MeV}]$')
plt.ylabel(r'$R(E_V, \bar{E}_V$)')
plt.legend()
plt.tight_layout()
plt.show()
```
